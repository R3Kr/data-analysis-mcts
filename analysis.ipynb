```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("scores.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df["TotalPoints"] = df["CIVIL"].add(df["MILITARY"]).add(df["SCIENCE"]).add(df["TRADE"]).add(df["GUILD"]).add(df["WONDER"]).add(df["GOLD"])
df.head()
```

```python
# Creating a boxplot
df.boxplot(column=['TotalPoints'], by='playerindex')

# Showing the plot
plt.title('Boxplot of TotalPoints by playerindex')
plt.suptitle('')  # Suppress the automatic title to avoid duplication
plt.ylabel('TotalPoints')
plt.show()
```

```python
mctsdf = df[df["playerindex"] == 0]
mctsdf.head()
```

```python
# Creating a boxplot
mctsdf.boxplot(column=['TotalPoints'], by='playouts')

# Showing the plot
plt.title('Boxplot of TotalPoints by playouts')
plt.suptitle('')  # Suppress the automatic title to avoid duplication
plt.ylabel('TotalPoints')
plt.show()
```

```python
df["won"] = [True if i % 3 == 0 else False for i in range(df["playerindex"].size)]
df.head(200)
```

```python
winrate_table = df[df["playerindex"] == 0]
# Group by 'Game Mode' and 'Map' and calculate winrate
winrate = winrate_table.groupby(['playouts'])['won'].mean()
winrate.head()

```

```python
movesdf = pd.read_csv("playedmoves.csv")
movesdf.head()
```

```python
movesdf["discarded"] = movesdf["movetype"].str.contains("DISCARD")
movesdf.groupby(["playerindex"])["discarded"].mean()
```

```python
mctsmoves = movesdf[movesdf["playerindex"] == 0]
mctsmoves["playouts"] = movesdf["gameid"].map(lambda x: 10 if (x < 100) else 20 if (x < 200) else 40 if (x < 300) else 80 if (x < 400) else 160)

mctsmoves.groupby(["playouts"])["discarded"].mean()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("scores.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df["TotalPoints"] = df["CIVIL"].add(df["MILITARY"]).add(df["SCIENCE"]).add(df["TRADE"]).add(df["GUILD"]).add(df["WONDER"]).add(df["GOLD"])
df.head()
```

```python
# Creating a boxplot
df.boxplot(column=['TotalPoints'], by='playerindex')

# Showing the plot
plt.title('Boxplot of TotalPoints by playerindex')
plt.suptitle('')  # Suppress the automatic title to avoid duplication
plt.ylabel('TotalPoints')
plt.show()
```

```python
mctsdf = df[df["playerindex"] == 0]
mctsdf.head()
```

```python
# Creating a boxplot
mctsdf.boxplot(column=['TotalPoints'], by='playouts')

# Showing the plot
plt.title('Boxplot of TotalPoints by playouts')
plt.suptitle('')  # Suppress the automatic title to avoid duplication
plt.ylabel('TotalPoints')
plt.show()
```

```python
df["won"] = [True if i % 3 == 0 else False for i in range(df["playerindex"].size)]
df.head(200)
```

```python
winrate_table = df[df["playerindex"] == 0]
# Group by 'Game Mode' and 'Map' and calculate winrate
winrate = winrate_table.groupby(['playouts'])['won'].mean().reset_index()

# Create a figure and an axes instance
fig, ax = plt.subplots()

# Hide the axes
ax.axis('off')

# Add table
table = ax.table(cellText=winrate.values, colLabels=["Playouts per turn", "Winrate"], loc='center', cellLoc='center')

# Adjust table properties
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  # you can change the size as needed

# Adjust layout to make room for the table
fig.tight_layout()

plt.show()
```

```python
movesdf = pd.read_csv("playedmoves.csv")
movesdf.head()
```

```python
movesdf["discarded"] = movesdf["movetype"].str.contains("DISCARD")
movesdf.groupby(["playerindex"])["discarded"].mean()
```

```python
mctsmoves = movesdf[movesdf["playerindex"] == 0]
mctsmoves["playouts"] = movesdf["gameid"].map(lambda x: 10 if (x < 100) else 20 if (x < 200) else 40 if (x < 300) else 80 if (x < 400) else 160 if (x < 500) else 320)

t = mctsmoves.groupby(["playouts"])["discarded"].mean().reset_index()

# Create a figure and an axes instance
fig, ax = plt.subplots()

# Hide the axes
ax.axis('off')

# Add table
table = ax.table(cellText=t.values, colLabels=["Playouts per turn", "Discardrate"], loc='center', cellLoc='center')

# Adjust table properties
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  # you can change the size as needed

# Adjust layout to make room for the table
fig.tight_layout()

plt.show()
```

```python
movetype_count_per_playouts = mctsmoves.groupby(["playouts"])["movetype"].value_counts().unstack()

# Setup for multiple pie charts
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 6))  # Adjust grid dimensions as needed
axes = axes.flatten()  # Flatten the grid for easy iteration

# Plot a pie chart for each 'Number'
for i, (idx, row) in enumerate(movetype_count_per_playouts.iterrows()):
    row.plot(kind='pie', ax=axes[i], autopct='%1.1f%%', startangle=90, title=f'Number of playouts: {idx}')
    axes[i].set_ylabel('')  # Remove the y-label

plt.tight_layout()
plt.show()
```
